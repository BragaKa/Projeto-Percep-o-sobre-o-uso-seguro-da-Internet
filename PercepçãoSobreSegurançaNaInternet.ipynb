## Gráfico

Nosso gráfico compara as respostas de pais e responsáveis com as de crianças e adolescentes, evidenciando diferenças no nível de conhecimento sobre segurança virtual. De modo geral, os responsáveis demonstraram maior conhecimento sobre as prevenções na internet, exceto em aspectos específicos, como proteção contra vírus e gerenciamento de senhas, nos quais as crianças se sobressaíram. Esse resultado sugere que grande parte do conhecimento das crianças sobre segurança digital é influenciada diretamente pelo nível de conscientização e orientação de seus responsáveis.

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os dados do arquivo XLSX
df = pd.read_excel("dados_detalhados_internet.xlsx", sheet_name="Sheet1")

# Renomear as colunas para facilitar o uso
df.columns = ["Resposta", "Pais ou Responsáveis (%)", "Crianças e adolescentes (%)"]

# Transformar os dados para o formato longo (ideal para Seaborn)
df_long = df.melt(id_vars="Resposta", var_name="Grupo", value_name="Porcentagem")

# Criar o gráfico de barras agrupadas
plt.figure(figsize=(12, 8))
sns.barplot(
    x="Resposta",
    y="Porcentagem",
    hue="Grupo",
    data=df_long,
    palette="coolwarm",  # Paleta de cores moderna
)

# Adicionar título e labels
plt.title("Comparação de Respostas entre Pais/Responsáveis e Crianças/Adolescentes", fontsize=16)
plt.xlabel("Respostas", fontsize=12)
plt.ylabel("Porcentagem (%)", fontsize=12)

# Rotacionar os rótulos do eixo X para melhorar a legibilidade
plt.xticks(rotation=45, ha="right")

# Ajustar a legenda
plt.legend(title="Grupo", bbox_to_anchor=(1.05, 1), loc='upper left')

# Melhorar a visualização dos eixos
plt.grid(True, linestyle='--', alpha=0.7, axis="y")

# Ajustar o layout para evitar cortes
plt.tight_layout()

# Mostrar o gráfico
plt.show()
```
